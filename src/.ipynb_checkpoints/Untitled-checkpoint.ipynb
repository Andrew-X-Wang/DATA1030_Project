### Pipeline?

```python
from sklearn.metrics import r2_score
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
```

```python
def MLpipe_KFold_RMSE(X_grouped, y, preprocessor, ML_algo, param_grid, metric, random_states=[]):
    '''
    This function splits the data to other/test (80/20) and then applies KFold with 4 folds to other.
    The RMSE is minimized in cross-validation.
    '''    
    test_size = 0.2
    train_size = 0.6
    n_splits = 4
#     metric = 'neg_root_mean_squared_error'
    
#     param_grid = ParameterGrid(param_grid)
    
    best_models = []
    test_scores = []
    for r_state in random_states:
        # other/test split
        X_train_grouped, X_test_grouped, y_train, y_test = train_test_split(X_grouped, y, 
                                                            test_size = test_size, 
                                                            random_state=r_state)
        print(X_train_grouped.shape)
        display(X_train_grouped)
        # group shuffle split
        group_train_split = GroupShuffleSplit(train_size = train_size/(1-test_size), random_state=r_state)
        
        # Removing group
        X_train, X_test = X_train_grouped[['R', 'C', 'time_step', 'u_in', 'u_out']], \
                          X_test_grouped[['R', 'C', 'time_step', 'u_in', 'u_out']]
#         X_test,  y_test  = X[['R', 'C', 'time_step', 'u_in', 'u_out']]].iloc[test_idxs],  y.iloc[test_idxs]
#         for train_idx, valid_idx in group_split_train.split(X, y, groups):
#             X_train, y_train = X.iloc[train_idxs], y.iloc[train_idxs]
#             X_valid, y_valid = X.iloc[valid_idxs], y.iloc[valid_idxs]

#             X_train_prep = feats_preprocessor.fit_transform(X_train)
#             X_val_prep = feats_preprocessor.transform(X_valid)
#             X_test_prep = feats_preprocessor.transform(X_test)

#             features_prep = list(feats_preprocessor.named_transformers_['cat'].get_feature_names(categorical_features)) + \
#                             list(feats_preprocessor.transformers_[1][-1]) + \
#                             list(feats_preprocessor.transformers_[2][-1])
       
        pipe = make_pipeline(preprocessor, ML_algo)
        
        # Gridsearch
        grid = GridSearchCV(pipe, param_grid=param_grid, scoring=metric, 
                            cv=group_train_split, return_train_score=True, 
                            verbose=True)
        grid.fit(X_train, y_train, groups=X_train_grouped[['breath_id']])
        
        best_model = grid.best_estimator_
        best_models.append(best_model)
        best_params = grid.best_params_

        # Best score on held-out data:
        y_pred = best_model.predict(X_test)
        y_pred = grid.predict(X_test)
        test_score = r2_score(y_test, y_pred)
        test_scores.append(test_score)
        
    return best_models, best_params, test_scores
```

```python
seed = 42
l2_linear = Ridge(random_state = seed)
n_params = 2
param_grid = {
    'ridge__alpha' : np.logspace(-4,2,n_params)
}
random_states = [100, 1000]
metric='r2'

best_ridge_models, best_ridge_params, l2_test_scores = MLpipe_KFold_RMSE(X_grouped, y, feats_preprocessor, l2_linear, param_grid, metric, random_states=random_states)
```
